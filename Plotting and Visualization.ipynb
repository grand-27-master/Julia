```julia
x=1:10;y=rand(10,2); #2 columns means 2 lines
plot(x,y,title="Two lines",label=["Line 1" "Line 2"],lw=2) #lw means line width
```

```julia
xlabel!("x-axis") # ! is used to add to an existing plot
ylabel!("y-axis")
```

```julia
scatter(x,y)
```

```julia
bar(x,y)
```

```julia
histogram(x,y)
```

```julia
pie(x,y)
```

```julia
#plotting all graphs in 1 window
p1=scatter(x,y,title="scatter plot")
p2=bar(x,y,title="bar plot")
p3=histogram(x,y,title="histogram")
p4=pie(x,y,title="pie chart")

plot(p1,p2,p3,p4,legend=false)
```
